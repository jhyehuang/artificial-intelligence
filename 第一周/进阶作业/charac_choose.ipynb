```python
data=data.drop('casual', axis = 1)     #删掉无关特征   casual
data=data.drop('registered', axis = 1) #删掉无关特征   registered
data=data.drop('instant', axis = 1)    #删掉无关特征   instant
data=data.drop('dteday', axis = 1)     #删掉无关特征   dteday
data_train = data[data.yr == 0].copy()       #获得训练集

yArry=data_train['cnt']
xArry=data_train.drop('cnt',axis = 1)
columns=xArry.columns
```

```python
from sklearn.metrics import r2_score  #评价回归预测模型的性能

def crossValidation(xArr,yArr,numVal=10):
    from sklearn.linear_model import  RidgeCV
    m = len(yArr) 
    alphas=[i/100 for i in range(1,10000)]
    ridge = RidgeCV(alphas=alphas,store_cv_values=True) 
    indexList = list(range(m))
    dest_model_score=0
    #dest_model=ridge
    columns=xArr.columns
    xArr=xArr.values
    for i in range(numVal):
        trainX=[]; trainY=[]
        testX = []; testY = []
        np.random.shuffle(indexList)
        for j in range(m):#create training set based on first 90% of values in indexList
            if j < m*0.9: 
                trainX.append(xArr[indexList[j]])
                trainY.append(yArr[indexList[j]])
            else:
                testX.append(xArr[indexList[j]])
                testY.append(yArr[indexList[j]])
        ridge.fit(trainX,trainY)    #get 30 weight vectors from ridge
        testY_pred=ridge.predict(testX)
        score=r2_score(testY, testY_pred)
        if dest_model_score < score:
            dest_model_score=score.copy()
            dest_model=ridge
    print ("the best model from Ridge Regression is:\n",dest_model_score)
    mse_mean = np.mean(dest_model.cv_values_, axis = 0)
    plt.plot(np.log10(alphas), mse_mean.reshape(len(alphas),1)) 

    #这是为了标出最佳参数的位置，不是必须

    plt.xlabel('log(alpha)')
    plt.ylabel('mse')
    plt.show()
    return dest_model,dest_model_score

    #print ("with constant term: ",-1*sum(multiply(meanX,unReg)) + np.mean(yMat))
```

```python
dest_model,dest_model_score=crossValidation(xArry,yArry,30)
```

```python
fs = pd.DataFrame({"columns":list(columns), "coef_ridge":list((dest_model.coef_.T))})
fs.sort_values(by=['coef_ridge'],ascending=False)
```
